```python
year = []
month = []
for date in df['date']:
  year.append(int(date[0:4]))
  month.append(int(date[4:6]))

df['year'] = year
df['month'] = month
```

```python
df.drop('date', axis=1)
```

```python
columns = df.columns.values.tolist()
y = df.loc[:, columns.pop(2)]
del columns[1]
X = df.loc[:, columns]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_tr, X_ts, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
scaler.fit(X_tr)

X_train = scaler.transform(X_tr)
X_test = scaler.transform(X_ts)
```

```python
print(X_train.shape, y_test.shape)
```

# MLP Model

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential([
     tf.keras.layers.Input(shape=(21,)),
     Dense(128, activation='relu'),
     Dense(32, activation='relu'),
     Dense(1)
])
# model.add(Dense(128, input_dim=X_train.shape[1], activation='relu'))
# model.add(Dense(32, activation='relu'))
# model.add(Dense(1))
```

```python
model.summary()
```

```python
from keras.optimizers import Adam, SGD
from keras.losses import MeanSquaredLogarithmicError, MeanSquaredError
```

```python
optimizers = ['rmsprop', 'adam']
losses = ['mean_squared_error', 'mean_squared_logarithmic_error']
```

```python
for lss in losses:
  for opt in optimizers:
    model.compile(loss=lss, optimizer=opt)
    history = model.fit(tf.expand_dims(X_train, axis=-1), y_train, validation_data=(X_test, y_test), epochs=160, verbose=1)
    train_lss = model.evaluate(X_train, y_train, verbose=0)

    test_lss = model.evaluate(X_test, y_test, verbose=0)
    print('loss function:', lss, 'oprimizer', opt)
    print('Train: %.3f, Test: %.3f' % (train_lss, test_lss))
    plt.title('Loss: %s & Optimizer %s' %(lss, opt))
    plt.plot(history.history['loss'], label='train')
    plt.plot(history.history['val_loss'], label='test')
    plt.legend()
    plt.show()
    print('\n\n')
```

# Prefered Model:
##    Loss Function: Mean Squared Logarithmic Error & Loss Function: rmsprop

```python
model.compile(loss='mean_squared_logarithmic_error', optimizer='rmsprop')
history = model.fit(tf.expand_dims(X_train, axis=-1), y_train, validation_data=(X_test, y_test), epochs=20, verbose=1)
train_lss = model.evaluate(X_train, y_train, verbose=0)
test_lss = model.evaluate(X_test, y_test, verbose=0)
print('Train: %.3f, Test: %.3f' % (train_lss, test_lss))
plt.title('Loss: %s & Optimizer %s' %(lss, opt))
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```
